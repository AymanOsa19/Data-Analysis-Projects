***Another Way For Visualizatoin***

```python
itemnames = olympix['Country'].value_counts().index[:10]
itemvalues = olympix['Country'].value_counts().values[:10]
plt.figure(figsize = (20 , 10))
plt.bar(itemnames , itemvalues , width = 0.7 , color = 'purple' , linewidth = 0.4)
plt.xlabel("Country" , fontsize = 'medium' )
plt.ylabel("Count" , fontsize = 'medium' )
plt.title('Number Of Medals for Each Country')
for x in range(len(itemvalues)) :
  plt.text(x, itemvalues[x] , itemvalues[x] , ha = 'center' , va = "bottom")
```

```python
itemnames = olympix['City'].value_counts().index[:10]
itemvalues = olympix['City'].value_counts().values[:10]
plt.figure(figsize = (20 , 10))
plt.bar(itemnames , itemvalues , width = 0.7 , color = 'olive' , linewidth = 0.4)
plt.xlabel("City" , fontsize = 'medium' )
plt.ylabel("Count" , fontsize = 'medium' )
plt.title('Number Of Medals for Each City')
for x in range(len(itemvalues)) :
  plt.text(x, itemvalues[x] , itemvalues[x] , ha = 'center' , va = "bottom")
```
